Tidying this dataset consisted of mostly dropping unwanted columns from our dataframe. We made sure to preserve the original dataset, since some columns we wanted to filter out for now but might be useful later. For example, the boolean datatypes such as drive-thru or mobile-order, might be beneficial later for comparing marketing strategies, but for now we are using this dataset to understand the quantity and location of Dunkin stores in the United States.

Next we wanted just one address row instead of two to comply with the attributes of tidy data we learned in class. Therefore we decided to join address_line_1 and address_line_2 using numpy. We used numpy so that we could exclude data under address_line_2 that was NaN, and append present data to address_line_1 in parenthesis. We considered re-ordering the hours for the days of the week, by creating a 'days' and 'hours' column. This did not work because it caused repetitions of our observations, and our purpose in this phase is to keep the location and row_id variables unique, therefore we decided to omit the hours and days of the week.

Finally we removed the row_id column and instead renamed the index 'row_id' to eliminate redundance.

```python
# Use .drop to exclude unwanted columns in our final dunkin dataframe. Made the row_id our primary index
final_dunkin= original_dunkin.drop(columns=['drive-thru', 'mobile-order', 'dunkin-card', 'kcup', 'curbside-pickup', 'has-baskin-robbins', 'row_id', 'mon_hrs','tue_hrs', 'wed_hrs', 'thu_hrs', 'fri_hrs', 'sat_hrs', 'sun_hrs'])
final_dunkin.index.name = 'row_id'

# Having two seperate columns for address line 1 and 2 was not tidy, use numpy to join non-empty address_line_2 names with address_line_1 in parenthesis to create a single address line column
final_dunkin['address'] = np.where(
    pd.notna(final_dunkin['address_line_2']),
    final_dunkin['address_line_1'] + ' (' + final_dunkin['address_line_2'] + ')',
    final_dunkin['address_line_1']
)
# Drop the previous address line columns since we replaced them with a single column
final_dunkin = final_dunkin.drop(columns=['address_line_1', 'address_line_2'])
```

### **City Population Statistics**

#### **Store Density Per City Capita**

```python
#COUNT STORES PER CITY
# Aggregate the raw store location data to get a single count per city/state pair

# Group Dunkin' data by the standardized 'city_name' and 'state_name' and count stores
dunkin_counts = final_dunkin.groupby(['city_name', 'state_name']).size().reset_index(name='Dunkin_Count')

# Do same for Starbucks stores
starbucks_counts = final_starbucks_df.groupby(['city_name', 'state_name']).size().reset_index(name='Starbucks_Count')

# Merge the two store counts together on the standardized keys: 'city_name' and 'state_name'
store_counts_df = pd.merge(dunkin_counts, starbucks_counts, on=['city_name', 'state_name'], how='outer').fillna(0)

# MERGE WITH POPULATION DATA
# To correctly inner join population data with store counts, we clean and standardize the population data

# To preserve the original Dataframe, create a copy of population data to work with
population_df = final_population_df.copy() #note: when copying final pop df, the standardized column names do not carry over

# Rename columns to standardized 'city_name' and 'state_name' (redo since overidden in copy)
# City -> city_name and STNAME -> state_name
population_df.rename(columns={'City': 'city_name', 'STNAME': 'state_name'}, inplace=True)

# *IMPORTANT*
# This ensures 'state_name' column that is currently full names, can be matched to the state_abbrev dict we made above
population_df['state_name'] = (
    population_df['state_name']
    .str.strip().str.title() # Strip extra white space, and convert to uppercase for clear formatting
    .map(state_to_abbrev) # Maps 'New York' --> 'NY'
    .str.upper() # Ensures uppercase to match with store data format
)

# Convert the population estimate of 2020 to numeric. Strings or nulls are coerced to NaN
population_df['POPESTIMATE2020'] = pd.to_numeric(population_df['POPESTIMATE2020'], errors='coerce')

# Final standardization of city names, stripping extra whitespace and ensuring uppercase
population_df['city_name'] = population_df['city_name'].astype(str).str.upper().str.strip()

# Drop NaNs after mapping to cleanup where standardization failed
population_df.dropna(subset=['city_name', 'state_name', 'POPESTIMATE2020'], inplace=True)


# INNER MERGE between store counts and the new tidied population data.
# We chose inner merge to only analyze cities that have complete store counts AND population data
city_analysis_df = pd.merge(
    store_counts_df,
    population_df,
    on=['city_name', 'state_name'],
    how='inner'
)
# Filter out records where population is too low (below 1000) so that our per capita calculation is stable
city_analysis_df = city_analysis_df[city_analysis_df['POPESTIMATE2020'] >= 1000]

# CALCULATE STORE DENSITY PER CAPITA (per 10,000 residents)
# Now we want to normalize store counts by population to create an accurate strore density metric for our y-axis

# Define the scale factor (per 10,000 residents)
POP_SCALE = 10000
# Calculate store density for each chain: store count/ population count
city_analysis_df['Dunkin_Density'] = (city_analysis_df['Dunkin_Count'] / city_analysis_df['POPESTIMATE2020']) * POP_SCALE
city_analysis_df['Starbucks_Density'] = (city_analysis_df['Starbucks_Count'] / city_analysis_df['POPESTIMATE2020']) * POP_SCALE

# CATEGORIZE CITIES INTO POPULATION BINS (very small, small, medium, large, very large)

# Define bins based on usual urban classification and choose labels
# Chosen bins are [0-10k], [10k-50k], [50k-250k], [250k-1M] and [1M+]
bins = [0, 10000, 50000, 250000, 1000000, np.inf]
labels = ['1k - 10k (Very Small)', '10k - 50k (Small)', '50k - 250k (Medium)', '250k - 1M (Large)', '1M+ (Very Large)']

# Use pd.cut() to assigne the population estimate into the defined bins above
city_analysis_df['Population_Bin'] = pd.cut(
    city_analysis_df['POPESTIMATE2020'],
    bins=bins,
    labels=labels,
    right=False, # Setting right=False makes the left bin inclusive and the right bin exclusive
    ordered=True # ordered=True preserves the order of catergories for aggregation and plotting
)

# FINAL AGGREGATION: AVERAGE DENSITY PER BIN

# Group the data using .groupby() with the new Population_Bin
# Calculate the mean density
density_by_bin = city_analysis_df.groupby('Population_Bin', observed=True)[['Dunkin_Density', 'Starbucks_Density']].mean().reset_index()

# Restructure the dataframe to 'long' format using melt.()
# This is crucial for plotting comparison bar charts with Seaborn, but we adapt it for Matplotlib
density_long_df = density_by_bin.melt(
    id_vars='Population_Bin',
    value_vars=['Dunkin_Density', 'Starbucks_Density'],
    var_name='Chain', # New column 'Chain' created to indentify which cahin the density belongs to
    value_name='Average_Density_Per_10k_People'
)

# Clean up Chain names (omitting "_Density")
density_long_df['Chain'] = density_long_df['Chain'].str.replace('_Density', '')

# CREATE VISUALIZATION

# Prepare data for plotting
bar_width = 0.35
population_bins = density_by_bin['Population_Bin']
x_indices = np.arange(len(population_bins)) # the x locations for the groups
dunkin_means = density_by_bin['Dunkin_Density']
starbucks_means = density_by_bin['Starbucks_Density']
plt.figure(figsize=(12, 7))

# Create bars for Dunkin'
plt.bar(
    x_indices - bar_width/2, # Shifted to the left
    dunkin_means,
    bar_width,
    label='Dunkin',
    color='#FF8000'
)

# Create bars for Starbucks
plt.bar(
    x_indices + bar_width/2, # Shifted to the right
    starbucks_means,
    bar_width,
    label='Starbucks',
    color='#00704A'
)

plt.title('Average Store Density Per Capita by City Population Size', fontsize=16)
plt.xlabel('City Population Bin (2020 Estimate)', fontsize=12)
plt.ylabel(f'Average Stores per {POP_SCALE:,} Residents', fontsize=12)
plt.xticks(x_indices, population_bins, rotation=0) # Set x-tick positions and labels
plt.legend(title='Coffee Chain')
plt.grid(axis='y', linestyle='--', alpha=0.7) # Add light horizontal gridlines
plt.tight_layout() # matplotlib function to adjust parameters for a tight and clean layout
plt.show()

# Output the numerical results for future use while modeling
print("Summary Table of Average Store Density per 10,000 People by City Size")
print(density_by_bin)
```

This statistic measures the average store density per 10,000 residents across five city population bins:very small, small, medium, large, and very large. The graph shows a difference in market saturation of the two stores, and confirms population cannot be ingored in future modeling.

#### **Model Methodology**

Our chosen methodology is Logistic Regression. We chose Logisitic over Linear Regression because our dependent variables are binary (Present/Absent) or categorical (Dominance, Competition), not continous numerical data which is required for Linear Regression.

*   Ensure all data is standardized


**2.   Scaling**   

This creates the necessary target variables and scales the predictors for stable model fitting.


*   Create target variables (Completed in EDA)
    *   Starbucks_Present, Dunkin_Present, Dominant_Brand
*   Scale Predictors
    *   Scale the continous, independent variables (Median Income and POPESTIMATE2020)

**3.   Binary Logistic Regression**

This is the first analytical stage, used to confirm the initial hypothesis observed in the Average Number of Stores Bar Plot.


*   **Model Fitting:** Fit two separate Binary Logistic Regression models, one for each chain's presence Starbucks_Present and Dunkin_Present, including Median as the main predictor and POPESTIMATE2020 and Founding_Region as controls.
*  **Test Hypothesis:** Use P-values to formally confirm the statistical significance of Median income. Use the Odds Ratio to quantify the directional change (positive for Starbucks, negative/neutral for Dunkin) in the odds of presence, independent of population.


**4.   Multinomial Logistic Regression**

This is the final analytical stage which provides the comprehensive market structure analysis required for Objective 2.


*   Model Fitting: Fit a single Multinomial Logistic Regression model using the Dominant_Brand target, comparing the odds of the three active market states (1, 2, 3) against the Neither baseline (0).
*   Comparative Inference: Utilize the separate Odds Ratios generated for each category to quantify:
